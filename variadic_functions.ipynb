 # Write a function to add any amount of numbers

# *args -> this will act like positional arguments which is considered tuple inside python

```python
def func1(*args):
    print(f"Arguments:{args}")
    print(f"Type:{type(args)}")

    for i in args:
        print (i)
```

```python
func1(2 ,4,5,8)
```

```python
func1(24, 25, "shubham", "yash", True)
```

```python
# How to unpack the list inside variadic function
# After unpacking individual example will be considered as inputs
b = (3, 4, 5, "shubham", "yash",False ,14,15)
func1(*b)
```

# Add any amount off numbers

```python
def add_multiple(*args):
    # initialize sum with 0
    s = 0
    # Apply for loop on args
    for i in args:
        s = s+i
    # Return the value of s
    return s    
```

```python
add_multiple(2,5,8,9,8,7,6)
```

```python
b =list(range(1,100))
add_multiple(*b)
```

# keyword arguments
**kwargs -> Act like Dictionary in python

```python
def func2(**kwargs):
    print(kwargs)
    print(type(kwargs))
    for key , value in kwargs.items():
        print(f"Key :{key}, Value:{value}")
```

```python
func2(name="Shubham", age=24)
```

```python
func2(name="Shubham",age=24,email= "shubhamdute@gmail.com")
```

# I want store user information with some flexibility

```python
def user_info(**kwargs):
    user ={
        "name":None,
        "age":None,
        "email":None
         }
    user.update(kwargs)
    return user
```

```python
a1 = user_info(name="shubham", age=24, email="shubhamdute@gmail.com")
print(a1)
```

```python
a2 = user_info(name="shubham",age=24)
print(a2)
```

```python
a3 = user_info(full_name= "Shubham Dute", age = 24, email= "shubhamdute@gmail.com")
print(a3)
```

```python
a4 = user_info(
    name = "Shubham",
    email = "Shubhamdute@gmail.com",
    city = "kotul",
    State = "maharashtra",
    married = False
)
print(a4)
```

# Perform retaurant billing
menu -> burger, pizza, pasta,coffe tea

```python
# MenU Prices
Menu_Prices= {
    "burger":134,
    "pizza":145,
    "pasta":150,
    "sandwich":120,
    "coffee":80,
    "tea":30

}
```

```python
# kwargs burger=2 ,menu, qty

def billing(**kwargs):
    # iniatilize total bill to 0 
    total_bill = 0
    # Apply for loop on kwargs
    for food, qty in kwargs.items():
        price = Menu_Prices.get(food)
        if price is not None:
            line_total = price * qty
            print(f"Food Item : {food}, Quantity : {qty}, Price : {price}, line Total : {line_total}")
            total_bill = total_bill + line_total
        else:
            print(f"Food item : {food} not available in menu")

     # print the final bill
    print(f"Total bill : {total_bill:2f} INR")    
    # return the total bill
    return total_bill


```

```python
b1 = billing( burger=2, pasta=3, sandwich=1, coffee=1)
```

# You can combine both postional and kwargs
1. positional args are always first
2. keyword args are last

```python
def fun3(*args, **kwargs):
    print(f"Args : {args},Type : {type(args)}")
    for i in args:
        print(i)
    print(f"Kwargs : {kwargs}, Type: {type(kwargs)}") 
    for key, value in kwargs.items():
        print(key, value)
```

```python
fun3(3 , 4, 5, 6, 8, 8)
```

```python
fun3(a="Example", b=23, c=45.12, d=0)
```

```python
fun3(3 , 4, 5, 6, name="Rahul", age=24,married=True)
```
