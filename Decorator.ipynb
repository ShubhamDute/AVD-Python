# Write a decorator function to welcome and thank the use

```python
def welcome(func):
    def wrapper(*args, **kwargs):
        print("Welcome User!")
        res = func(*args, **kwargs)
        print(f"Results : {res}")
        print("Thank you")
        return res
        
    return wrapper
```

```python
# simple intrest function

@welcome
def simple_intrest (p, n, r):
    i = (p * n * r) / 100
    a = p + i
    return i , a
```

```python
i1 , a1 = simple_intrest(p=50000, n=5, r=7.5)
```

# write a decorator function to check execution time

```python
import time
```

```python
start = time.perf_counter()
time.sleep(2)
print("Hello")
time.sleep(3)
print("World")
stop = time.perf_counter()
elapsed_time = stop - start
print(f"Elapsed Time : {elapsed_time: .2f} sec")
```

```python
def time_decorator(func):
    def wrapper(*args , **kwargs):
        start = time.perf_counter()
        res = func(*args, **kwargs)
        print(f"Result : {res}")
        stop = time.perf_counter()
        elapsed_time = stop - start
        print(f"Elapsed time : {elapsed_time :.2f} sec")
        return res
    return wrapper
```

```python
@time_decorator
def square(n: float):
    time.sleep(3)
    return n**2
```

```python
s1 = square(12.5)
```

```python
s2 = square(20.5)
```

```python
@time_decorator
def square_multiple(nums: list[float]) -> list[float]:
    s = []
    for i in nums:
        time.sleep(1)
        print(f"Square of number {i} is {i**2}")
        s.append(i**2)
    return s
```

```python
s3 = square_multiple(nums = [12, 13, 32, 22, 23])
```

```python
s4 = square_multiple(nums= [22, 44, 55, 66, 77, 12, 13, 3, 11, 24])
```
